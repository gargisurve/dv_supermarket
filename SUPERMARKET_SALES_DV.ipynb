```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

**UNIVARIATE ANALYSIS**

```python
# 1. Are there any outliers in terms of total sales and unit price?
# BOX PLOT
plt.figure(figsize=(7,6))
plt.subplot(2,2,1)
sns.boxplot(y='Total', data=df, orient='vertical')
plt.title('Outliers in Total Sales')
plt.subplot(2,2,2)
sns.boxplot(y='Unit price', data=df, orient='vertical')
plt.title('Outliers in Unit Prices')
plt.tight_layout()
```

```python
# 2. What is the distribution of customer ratings?
# HISTOGRAM
sns.histplot(data=df, x='Rating', bins=7, kde=True, color='orange', edgecolor='blue')
plt.xlabel('Customer Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Ratings')
plt.show()
```

```python
# 3. What is the gender distribution of customers?
# PIE CHART [DIRECT PLOTTING]
gender_distribution = df['Gender'].value_counts()
gender_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution of Customers')
plt.show()
```

```python
# 4. What is the distribution of unit prices?
# HISTOGRAM
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Unit price', bins=20, kde=True, color='green', edgecolor='blue')
plt.title("Distribution of Unit Prices")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

**BIVARIATE ANALYSIS**

```python
# 5. Which city has the highest sales?
# PIE CHART
total_sales_by_city = df.groupby('City')['Total'].sum().sort_values(ascending=False)
e=[0.2,0.0,0.0]
plt.figure(figsize=(5, 5))
plt.pie(total_sales_by_city, labels=total_sales_by_city.index, autopct='%1.1f%%', startangle=140, explode=e, shadow=True)
plt.title("Total Sales by City",)
plt.show()
```

```python
# 6. What is the product category with the highest average rating?
# BAR PLOT
avg_rating_by_category = df.groupby('Product line')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_rating_by_category.index, y=avg_rating_by_category.values, palette='rocket')
plt.title("Average Rating by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=50)
plt.show()
```

```python
# 7. Which is the most preferred payment method?
# BAR PLOT
payment_method_counts = df['Payment'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=payment_method_counts.index, y=payment_method_counts.values, palette='viridis')
plt.title("Payment Method Preferences")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. What is the relationship between quantity and total sales?
# SCATTER PLOT
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Quantity', y='Total', color='orange')
plt.title("Quantity vs. Total Sales")
plt.xlabel("Quantity")
plt.ylabel("Total Sales")
plt.show()
```

**MULTIVARIATE ANALYSIS**

```python
# 9. How does the total sales amount vary across branches for each customer type?
# BAR PLOT
br = df['Branch'].sort_values(ascending=True)
plt.figure(figsize=(7, 6))
sns.barplot(x=br, y='Total', hue='Customer type', data=df, ci=None, palette='Set1')
plt.xlabel('Branch')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Branch and Customer Type')
plt.legend(loc='upper left', ncol=2)
plt.show()
```

```python
# 10. Calculate the correlation matrix between unit price, quantity, total sales and rating.
# HEATMAP
corr_matrix = df[['Unit price', 'Quantity', 'Total', 'Rating']]
corr_data = corr_matrix.corr()
# Generate a heatmap
plt.figure(figsize=(6, 6), dpi=100)
sns.heatmap(corr_data, annot=True, cmap='flare', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 11. How does the distribution of total sales vary between different product categories with respect to gender?
# VIOLIN PLOT
plt.figure(figsize=(7, 6))
sns.violinplot(data=df, x='Product line', y='Total', hue='Gender', palette="husl")
plt.title("Total Sales Distribution by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.legend(loc='upper left', ncol=2)
plt.show()
```
